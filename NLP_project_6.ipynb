```python
import kagglehub

# Download dataset
path = kagglehub.dataset_download("amananandrai/ag-news-classification-dataset")

print("Dataset downloaded to:", path)
```

```python
import pandas as pd
import os
```

```python
# Show files inside the downloaded folder
print(os.listdir(path))
```

```python
train = pd.read_csv(path + "/train.csv")
test = pd.read_csv(path + "/test.csv")
```

```python
import os

for root, dirs, files in os.walk(path):
    for file in files:
        print(os.path.join(root, file))
```

```python
train.columns = ["Category", "Title", "Description"]
test.columns = ["Category", "Title", "Description"]

train.head()
```

```python
train["Text"] = train["Title"] + " " + train["Description"]
test["Text"] = test["Title"] + " " + test["Description"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
tfidf = TfidfVectorizer(stop_words='english')

X_train = tfidf.fit_transform(train["Text"])
X_test = tfidf.transform(test["Text"])

y_train = train["Category"]
y_test = test["Category"]
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, prediction)

print("Accuracy:", accuracy)
```

```python
news = ["India won the cricket world cup after defeating Australia."]

news_vector = tfidf.transform(news)

result = model.predict(news_vector)

print("Predicted Category:", result[0])
```
