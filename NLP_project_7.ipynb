```python
# Import KaggleHub
import kagglehub

# Download the dataset
dataset_path = kagglehub.dataset_download("andrewmvd/trip-advisor-hotel-reviews")

print("Dataset downloaded successfully!")
print(dataset_path)
```

```python
import pandas as pd
import os
```

```python
# Check the downloaded files
print(os.listdir(dataset_path))
```

```python
reviews = pd.read_csv(dataset_path + "/tripadvisor_hotel_reviews.csv")

reviews.head()
```

```python
# Display basic information
reviews.info()

# Check for missing values
reviews.isnull().sum()
```

```python
# Convert ratings into sentiments

def find_sentiment(rating):
    if rating >= 4:
        return "Positive"
    elif rating <= 2:
        return "Negative"
    else:
        return "Neutral"

reviews["Sentiment"] = reviews["Rating"].apply(find_sentiment)

reviews.head()
```

```python
# Select review text
x = reviews["Review"]

# Select sentiment labels
y = reviews["Sentiment"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words="english")

x = vectorizer.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()

model.fit(x_train, y_train)

print("Model trained successfully!")
```

```python
prediction = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, prediction)

print("Accuracy :", round(accuracy * 100, 2), "%")
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, prediction))
```

```python
my_review = [
    "The room was very clean and the staff were friendly."
]

review_vector = vectorizer.transform(my_review)

result = model.predict(review_vector)

print("Predicted Sentiment :", result[0])
```

```python
my_review = [
    "The room was dirty and the service was terrible."
]

review_vector = vectorizer.transform(my_review)

result = model.predict(review_vector)

print("Predicted Sentiment :", result[0])
```
